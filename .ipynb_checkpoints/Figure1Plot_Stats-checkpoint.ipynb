```python
result = smf.ols(formula="Coef ~ GAL4*Behavior+GAL4*RegionNames+Behavior*RegionNames+UAS+FR", data=dfWGF).fit()
aov_table = anova_lm(result, typ=2)
print(result.rsquared)
aov_table
```

```python
result = smf.ols(formula="R2 ~ Behavior*RegionNames+GAL4+UAS+FR", data=dfWGF).fit()
aov_table = anova_lm(result, typ=2)
print(result.rsquared)
aov_table
```

```python
from patsy import dmatrices
y, X = dmatrices('R2 ~ Behavior*RegionNames+GAL4+UAS+FR', dfWGF, return_type='dataframe')
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif.round(1)
```
